8. Drop columns with more than 50% missing values. 

```python
missing_percentage = df.isnull().mean() * 100
df = df.loc[:, missing_percentage <= 50]
df
```

9. Fill missing Club values with "Free Agent". 

```python
df['Club Name'] = df['Club Name'].fillna('Free Agent')
```

10. Replace missing Height values with the median height. 

```python
df['Height(in cm)']=df['Height(in cm)'].fillna(df['Height(in cm)'].median())
```

11.Create one columns with null & non null

```python
import numpy as np
#how many rows
n=len(df)
non_ratio=0.3
num_nan=int(n*non_ratio)
values=np.random.randint(1,101,n).astype(float)
values[:num_nan]=np.nan
np.random.shuffle(values)
df['RandomScore']=values
df['RandomScore']
```

12. Replace missing RandomScore values with the median RandomScore.

```python
df
```

```python
def fill_height(df):
    if 'RandomScore' in df.columns:
        df['RandomScore']=df['RandomScore'].fillna(df['RandomScore'].median())
        print('median filled successfully')
    else:
        print("Column 'RandomScore' does not exist in the DataFrame.",df.columns)
    return df
fill_height(df)
```

Advanced Operations 

24.Create a new column BMI using Weight and Height. 


```python
df['BMI']= df['Weight(in kg)'] / ((df['Height(in cm)'] / 100) ** 2)
df[['Full Name','Weight(in kg)','Height(in cm)','BMI']]
```

25.Apply a function to categorize players into Low (<70), Medium (70- 80), High (>80) based on Overall. 

```python
def categorize(x):
    if x<70:
        return 'low'
    elif 70<x<80:
        return 'medium'
    else:
        return 'high'
df['Overall_Category'] = df['Overall'].apply(categorize)
print(df[['Full Name', 'Overall', 'Overall_Category']].head())
```

26.Use pd.cut() to bin players into 5 equal-sized groups based on Wage. 

```python
df['Wage_Group'] = pd.cut(df['Wage(in Euro)'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
print("Wage Group added. First few rows:")
print(df[['Full Name', 'Wage(in Euro)', 'Wage_Group']].head())
```

27.Calculate the average Potential by Age Group (e.g., U20, 20-30, 30+). 

```python
bins=[0,20,30,100]
labels=['U20','20-30','30+']
df['Age_Group']=pd.cut(df['Age'],bins=bins,labels=labels)
df['Age_Group']
avg_potential_by_age=df.groupby('Age_Group')['Potential'].mean().round(1)
avg_potential_by_age
```

28.Use pivot_table() to show average Overall by Position and Preferred Foot.

34.Plot a histogram of player Overall ratings. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,6))
sns.histplot(df['Overall'],bins=20,kde=True,color='blue')
plt.title('Player Overall Ratings')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.show()
```

35.Create a bar plot of the top 10 nationalities by player count. 

```python
top_nationalities=df['Nationality'].value_counts().head(10)
plt.Figure(figsize=(10,6))
sns.barplot(x=top_nationalities.values, y=top_nationalities.index)
plt.title('Top 10 Nationalities by Player Count')
plt.xlabel('Number of Players')
plt.ylabel('Nationality')
plt.show()
```

36.Generate a box plot of Wage by Position.

38.Visualize the correlation matrix between Overall, Potential, and Wage.

```python
correlation_matrix = df[['Overall', 'Potential', 'Wage(in Euro)']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix: Overall, Potential, Wage')
plt.show()
```
